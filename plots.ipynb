```python
mts, mps, ss = [], [], []
xs = np.arange(10, 110, 10)
```

```python
for name in ['MT', 'MP', 'S']:
    for i in xs:
        with open(f'./Part2/{name}_{i}.txt', 'r') as file:
            data = file.read()
        if name == 'MT':
            mts.append(float(data.split('\n')[-3].split(":")[1].strip().split(" ")[0]))
        if name == 'MP':
            mps.append(float(data.split('\n')[-3].split(":")[1].strip().split(" ")[0]))
        if name == 'S':
            ss.append(float(data.split('\n')[-3].split(":")[1].strip().split(" ")[0]))
```

```python
mts1, mps1 = [], []
```

```python
for name in ['MT', 'MP']:
    for i in xs:
        with open(f'./Part2/{name}_{i}.txt', 'r') as file:
            data = file.read()
        if name == 'MT':
            mts1.append(float(data.split('\n')[-3].split(":")[1].strip().split(" ")[0]))
        if name == 'MP':
            mps1.append(float(data.split('\n')[-3].split(":")[1].strip().split(" ")[0]))
```

```python
mts2, mps2 = [], []
```

```python
for name in ['MT', 'MP']:
    for i in xs:
        with open(f'./Part2/{name}_{i}.txt', 'r') as file:
            data = file.read()
        if name == 'MT':
            mts2.append(float(data.split('\n')[-3].split(":")[1].strip().split(" ")[0]))
        if name == 'MP':
            mps2.append(float(data.split('\n')[-3].split(":")[1].strip().split(" ")[0]))
```

```python
np.savetxt('./Part2/ss.txt', np.array(ss), fmt='%.6f')
np.savetxt('./Part2/mp.txt', mp, fmt='%.6f')
np.savetxt('./Part2/mt.txt', mt, fmt='%.6f')
```

```python
# median of mp and mt along the 3 rows
mp_median = np.median(mp, axis=0)
mt_median = np.median(mt, axis=0)

mp_median, mt_median
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))
plt.plot(xs, mp_median, marker='D', color='b', label='Multi-Processing Server')
plt.plot(xs, mt_median, marker='o', color='g', label='Multi-Threading Server')
plt.xlabel('Number of Client Processes')
plt.ylabel('Execution Time (s)')
plt.legend()
plt.title('Median of MP and MT')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
plt.plot(xs, np.array(ss), marker='o', color='red', label='Single Process Server')
plt.xlabel('Number of Client Processes')
plt.ylabel('Execution Time (s)')
plt.legend()
plt.title('Single Process Server Time')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))
plt.plot(xs, mp_median, marker='D', color='b', label='Multi-Processing Server')
plt.plot(xs, mt_median, marker='o', color='g', label='Multi-Threading Server')
plt.plot(xs, np.array(ss), marker='o', color='red', label='Single Process Server')
plt.xlabel('Number of Client Processes')
plt.ylabel('Execution Time (s)')
plt.yscale('log')
plt.legend()
plt.title('Execution Time for Different Server Designs (Log Scale)')
plt.show()
```
